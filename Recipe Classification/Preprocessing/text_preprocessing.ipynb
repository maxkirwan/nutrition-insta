```python
def preprocess_text(text):
    
    # Removing all @mentions
    clean_text = re.sub("@[A-Za-z0-9_]+","", text)
    
    # Removing all #hashtags
    clean_text = re.sub("#[A-Za-z0-9_]+","", clean_text)
    
    # Removing all emojis
    clean_text = demoji.replace(clean_text)
    
    # Replacing unicode fractions with whitespace + ascii representations
    fractions = {
        '\x215b': '1/8',
        '\x215c': '3/8',
        '\x215d': '5/8',
        '\x215e': '7/8',
        '\x2159': '1/6',
        '\x215a': '5/6',
        '\x2155': '1/5',
        '\x2156': '2/5',
        '\x2157': '3/5',
        '\x2158': '4/5',
        '\xbc': ' 1/4',
        '\xbe': '3/4',
        '\x2153': '1/3',
        '\x2154': '2/3',
        '\xbd': '1/2'
        }
    for f_unicode, f_ascii in fractions.items():
        clean_text = clean_text.replace(f_unicode, ' ' + f_ascii)
        
    
    
    return clean_text
```
