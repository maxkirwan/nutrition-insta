# Rule based recipe classifier

This notebook is my first attempt at classifying recipes from Instagram post descriptions.

Here I will attempt to identify recipes using rule-based methods.

```python
import numpy as np
import pandas as pd
```

```python
posts = pd.read_csv("/Users/maxkirwan/Desktop/Uni/Data Science MSc/Data Science Project/nutrition-insta/Instagram Data Scraping/Phantom Buster/delicouslyella_posts.csv")
posts.drop_duplicates(subset="postUrl", keep='first', inplace=True)
posts
```

```python
# simple function to determine whether post is a recipe
def is_recipe(text):
    text = str(text)
    if text.count('\n') > 8:
        return True
    else:
        return False
```

```python
# make new column stating whether the post is a recipe or not
posts["isRecipe"] = posts.description.apply(is_recipe)
```

```python
recipe_posts = posts[posts.isRecipe]
# print for more readable format
for i in range(len(recipe_posts)):
    print(recipe_posts.description.iloc[i])
    print("\n--------\n")
```
