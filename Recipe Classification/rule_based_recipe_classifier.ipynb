# Rule based recipe classifier

This notebook is my first attempt at classifying recipes from Instagram post descriptions.

Here I will attempt to identify recipes using rule-based methods.

```python
# simple function to determine whether post is a recipe
def is_recipe(text):
    text = str(text)
    if text.count('\n') > 8:
        return True
    else:
        return False
```

```python
# make new column stating whether the post is a recipe or not
posts["isRecipe"] = posts.description.apply(is_recipe)
```

```python
recipe_posts = posts[posts.isRecipe]
# print for more readable format
for i in range(len(recipe_posts)):
    print(recipe_posts.description.iloc[i])
    print("\n--------\n")
```

### Rule based function

Writing a rule based function to capture as many posts as possible. It's not going to be perfect, but at least it should get rid of a good chunk of the "noise" (i.e. any posts which clearly do not have an ingredients list).

```python
def includes_ingredient_list(text):
    
    # Initial assumption is to be false
    binary = False
    
    if "ingredients" in text:
        binary = True
        
    text_str = str(text)
    if text_str.count('\n') > 8:
        binary = True
        
    return binary
```

```python
# Adding includes_ingredient_list binary column
posts['includes_ingredient_list'] = posts['description_preprocessed'].apply(includes_ingredient_list)
```

```python
for post in posts[posts['includes_ingredient_list']]['description_preprocessed']:
    print(post)
    print("\n----------\n")
```
