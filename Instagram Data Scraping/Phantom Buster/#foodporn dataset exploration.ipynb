# Phantom Buster - Instagram hashtag collector

Used Phantom Buster to scrape 4773 posts with the hashtag #foodporn


This notebook explores these posts.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_csv("#foodporn.csv")
df.head(3)
```

```python
df.shape
```

```python
# Sort df by most likes
df.sort_values(by='likeCount', ascending=False, inplace=True)
```

```python
df.head(2)
```

```python
def get_list_of_hashtags(description):
    return re.findall('#[A-Za-z]+', str(description))

df['hashtags'] = df.apply(lambda row: get_list_of_hashtags(row['description']), axis=1)
```

```python
df
```

```python
df.description[0]
```

### Is there any correlation between the number of hashtags used and the number of likes received

```python
# Looks like there is no correlation between the number of hashtags used and the number of likes received per post

plt.figure(figsize=(10,8))
plt.scatter(x=[len(hashtag_list) for hashtag_list in df.head(1000).hashtags], y=df.head(1000).likeCount)
plt.ylabel('Number of likes')
plt.xlabel('Number of hashtags')
plt.show()
```

### What is more popular... images or videos?

```python
df_images = df[df.videoUrl.isna()]
df_vids = df[df.imgUrl.isna()]

print(f'Number of image posts: {len(df_images)}')
print(f'Number of video posts: {len(df_vids)}')
```

```python
df_images.likeCount.mean(), df_vids.likeCount.mean()
```

### Relationship between comments and likes

```python
plt.figure(figsize=(8,6))
plt.scatter(x=df.head(1000).likeCount, y=df.head(1000).commentCount)
plt.xlabel('Number of likes')
plt.ylabel('Number of comments')
plt.show()
```
